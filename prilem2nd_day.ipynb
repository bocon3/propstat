```python
import pandas as pd
from google.colab import files

# Upload CSV
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)

print("SUS Dataset")
print(df)

# SUS Question Columns
sus_columns = [f"Q{i}" for i in range(1, 11)]
print("\nSUS Columns:", sus_columns)

# Calculate SUS Score
def calculate_sus_score(row):
    score = 0

    for i, col in enumerate(sus_columns, start=1):
        response = row[col]

        # Odd-numbered questions
        if i % 2 == 1:
            score += response - 1

        # Even-numbered questions
        else:
            score += 5 - response

    return score * 2.5

# Compute SUS scores
df["sus_score"] = df.apply(calculate_sus_score, axis=1)

# Display computed SUS scores
print("\nIndividual SUS Scores")
if "user" in df.columns:
    print(df[["user", "sus_score"]])
else:
    print(df[["sus_score"]])

#descriptive statistics

mean_sus = df['sus_score'].mean()
median_sus = df['sus_score'].median()
var_sus = df['sus_score'].var()
std_sus = df['sus_score'].std()
cv_sus = (std_sus / mean_sus) * 100


print("\nDescriptive Statistics")
print(f" mean{mean_sus:.3f}")
print(f" median{median_sus:.3f}")
print(f"variance {var_sus:.3f}")
print(f" standard deviation{std_sus:.3f}")
print(f"coeffecient variation {cv_sus:.3f}")

#sus interpretation

#Acceptability
if mean_sus >=  72.5:
  acceptability="Acceptable"
elif mean_sus >= 51.7:
  acceptibility = "marginal"
else:
  acceptability="not acceptable"
  print(f"acceptability:{acceptability}")

#  ADJECTIVE RATING
if mean_sus >= 80.3:
   adj_rating="Excellent"
elif mean_sus >= 68.1:
  adj_rating="Good"
elif mean_sus >= 68:
 adj_rating="Okay"
elif mean_sus >= 51:
 adj_rating="Poor"
else:
  adj_rating="Awful"
  print(f"acceptability:{adj_rating}")

```
